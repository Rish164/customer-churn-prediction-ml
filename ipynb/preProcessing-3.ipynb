```python
import pandas as pd
import numpy as np

df = pd.read_csv("../dataset/cleaned_churn.csv")

#Target Variable Mapping, this is not called Encoding, this is just mapping the target variable to 0 and 1.
#We encode features, we map target variable, both are different things.
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

df.head()
df.isnull().sum()
```

```python
#Feature Separation

x = df.drop("Churn", axis=1)
y = df["Churn"]

y.to_csv("../dataset/target.csv", index=False)
```

```python
#Encoding Categorical Variables

#We will use get_dummies for encoding categorical variables, it will create new columns for each category and assign 1 or 0 based on the presence of that category in the original column.
#This is called One-Hot Encoding, it is a common method for encoding categorical variables.

x = pd.get_dummies(x, drop_first=True)

#Shape before encoding
print(df.shape)

#Shape after encoding
print(x.shape)

x.isnull().sum()
x.head()
x.dtypes
```

```python
bool_cols = x.select_dtypes(include='bool').columns
x[bool_cols] = x[bool_cols].astype(int)
x.head()
print(x.dtypes)

#Although the StandardScaler can handle boolean values, it is generally recommended to convert them to integers (0 and 1) before scaling, as it can help improve the performance of the model and ensure that the scaling is applied correctly. By converting boolean values to integers, we can ensure that the StandardScaler treats them as numerical features and scales them appropriately.
```

```python
#Scaling Numerical Features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
#This gives us a numpy array, we can convert it back to a dataframe

x_scaled = pd.DataFrame(x_scaled, columns=x.columns)

x_scaled.head()

x_scaled.to_csv("../dataset/preprocessed_churn.csv", index=False)

#Dataset is now preprocessed and ready for modeling.
```
