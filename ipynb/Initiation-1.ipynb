```python
#Dataset is imbalanced...
#This becomes a problem because the model will be biased towards the majority class and will not perform well on the minority class.
#We just need to know this information for now, we will deal with it later.
df.dtypes
```

If the dataset will be biased on the majority class, the model would obviously predict results based on that,
which would lead to a high accuracy but low precision and recall for the minority class.

In short the model would be good at predicting the majority class but bad at predicting the minority class, which is not what we want.
For research purposes, we can use the imbalanced dataset to see how the model performs on it, but for real-world applications, we would need to address the imbalance issue to get better performance on the minority class.

Meaning: Balancing the dataset is important to predict churn accurately for both classes, and to avoid a biased model that performs well on the majority class but poorly on the minority class.
Because we don't want a model that just predicts the majority class, we want a model that can predict both classes accurately.
In this case, we want a model that can detect churners (minority class) as well as non-churners (majority class) accurately, so we need to address the imbalance issue to achieve that.

This approach is important for real-world business applications, where accurately identifying churners can help in customer retention strategies, while also maintaining good performance on non-churners to ensure overall model effectiveness.

We don't balance the data because it's wrong.
We balance data because our objective is to build a model that can predict both classes accurately, and to do that we need to address the imbalance issue.
#Because detecting churners is more important for a company than detecting non-churners, we need to ensure that our model can predict churners accurately, which is why we need to balance the dataset.
